### 60. How do you create a custom aggregation function to calculate the weighted average 'Salary' by 'Department'?

```python
# Assuming weight as AGE coloumn
df.groupby('DEPARTMENT').agg(
    Weighted_Avg_Salary=pd.NamedAgg(
        column='SALARY', 
        aggfunc=lambda x: (x * df.loc[x.index, 'AGE']).sum() / df.loc[x.index, 'AGE'].sum()
    )
).reset_index()
```
